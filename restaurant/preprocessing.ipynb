```python
rest=pd.read_csv('csv_files/rest.csv')
rest.fillna('', inplace=True)
rest.head()
```

```python
def jac_qgm_3_qgm_3(seq1, seq2):
    qgm1 = []
    qgm2 = []
    seq1 = '##' + seq1 + '$$'
    seq2 = '##' + seq2 + '$$'
    for i in range(len(seq1) - 2):
        qgm1.append(seq1[i:i + 3])
    for i in range(len(seq2) - 2):
        qgm2.append(seq2[i:i + 3])
    qgm1 = set(qgm1)
    qgm2 = set(qgm2)
    return len(qgm1 & qgm2) / len(qgm1 | qgm2)
```

```python
hhh=open('csv_files/hhh.csv','w')
hhh.write('_id,id1,id2,name_name_jac_qgm_3_qgm_3,address_address_jac_qgm_3_qgm_3,city_city_jac_qgm_3_qgm_3,flavor_flavor_jac_qgm_3_qgm_3,gold\n')
_id=0
for index1,row1 in rest.iterrows():
    for index2,row2 in rest.iterrows():
        if index1>=index2:
            continue
        id1,name1,address1,city1,flavor1,cluster_id1=row1['id'],row1['name'],row1['address'],row1['city'],row1['flavor'],row1['cluster_id']
        id2,name2,address2,city2,flavor2,cluster_id2=row2['id'],row2['name'],row2['address'],row2['city'],row2['flavor'],row2['cluster_id']
        name_name_jac_qgm_3_qgm_3=jac_qgm_3_qgm_3(name1,name2)
        address_address_jac_qgm_3_qgm_3=jac_qgm_3_qgm_3(address1,address2)
        city_city_jac_qgm_3_qgm_3=jac_qgm_3_qgm_3(city1,city2)
#         print(index1,index2,flavor1,flavor2)
        flavor_flavor_jac_qgm_3_qgm_3=jac_qgm_3_qgm_3(flavor1,flavor2)
        if cluster_id1==cluster_id2:
            hhh.write('{},{},{},{},{},{},{},{}\n'.format(_id,id1,id2,name_name_jac_qgm_3_qgm_3,address_address_jac_qgm_3_qgm_3,city_city_jac_qgm_3_qgm_3,flavor_flavor_jac_qgm_3_qgm_3,1))
            _id+=1
        else:
            hhh.write('{},{},{},{},{},{},{},{}\n'.format(_id,id1,id2,name_name_jac_qgm_3_qgm_3,address_address_jac_qgm_3_qgm_3,city_city_jac_qgm_3_qgm_3,flavor_flavor_jac_qgm_3_qgm_3,0))
            _id+=1
hhh.close()
```

```python
def blocking(row1, row2):
    name1,address1,city1,flavor1,cluster_id1=row1['name'],row1['address'],row1['city'],row1['flavor'],row1['cluster_id']
    name2,address2,city2,flavor2,cluster_id2=row2['name'],row2['address'],row2['city'],row2['flavor'],row2['cluster_id']
    
    # name should have overlap word
    name1_set=set(name1.split())
    name2_set=set(name2.split())
    if (len(name1_set)>1 or len(name2_set)>1) and ((name1_set & name2_set)==set()):
        return False
    
    return True
```

```python
hhh=open('csv_files/hhh_blocking.csv','w')
hhh.write('_id,id1,id2,name_name_jac_qgm_3_qgm_3,address_address_jac_qgm_3_qgm_3,city_city_jac_qgm_3_qgm_3,flavor_flavor_jac_qgm_3_qgm_3,gold\n')
_id=0
for index1,row1 in rest.iterrows():
    for index2,row2 in rest.iterrows():
        if index1>=index2:
            continue
        id1,name1,address1,city1,flavor1,cluster_id1=row1['id'],row1['name'],row1['address'],row1['city'],row1['flavor'],row1['cluster_id']
        id2,name2,address2,city2,flavor2,cluster_id2=row2['id'],row2['name'],row2['address'],row2['city'],row2['flavor'],row2['cluster_id']
        if blocking(row1, row2)==False:
            if cluster_id1==cluster_id2:
                print('error',id1,id2)
            continue
        name_name_jac_qgm_3_qgm_3=jac_qgm_3_qgm_3(name1,name2)
        address_address_jac_qgm_3_qgm_3=jac_qgm_3_qgm_3(address1,address2)
        city_city_jac_qgm_3_qgm_3=jac_qgm_3_qgm_3(city1,city2)
#         print(index1,index2,flavor1,flavor2)
        flavor_flavor_jac_qgm_3_qgm_3=jac_qgm_3_qgm_3(flavor1,flavor2)
        if cluster_id1==cluster_id2:
            hhh.write('{},{},{},{},{},{},{},{}\n'.format(_id,id1,id2,name_name_jac_qgm_3_qgm_3,address_address_jac_qgm_3_qgm_3,city_city_jac_qgm_3_qgm_3,flavor_flavor_jac_qgm_3_qgm_3,1))
            _id+=1
        else:
            hhh.write('{},{},{},{},{},{},{},{}\n'.format(_id,id1,id2,name_name_jac_qgm_3_qgm_3,address_address_jac_qgm_3_qgm_3,city_city_jac_qgm_3_qgm_3,flavor_flavor_jac_qgm_3_qgm_3,0))
            _id+=1
hhh.close()
```

```python
# Explore m_data

# Output the matched pairs whose name similarity is not 1
min_name_sim=1
for index, row in m_data.iterrows():
    if row['name_name_jac_qgm_3_qgm_3']<1:
        if row['name_name_jac_qgm_3_qgm_3']<min_name_sim:
            min_name_sim=row['name_name_jac_qgm_3_qgm_3']
        print('id1:{}, id2:{}, name1:{}, name2:{}, name_name_jac_qgm_3_qgm_3:{}'.format(row['id1'], row['id2'], rest['name'][row['id1']], rest['name'][row['id2']],row['name_name_jac_qgm_3_qgm_3']))
print('min_name_sim:{}\n'.format(min_name_sim))


# Output the matched pairs whose address similarity is not 1
min_address_sim=1
for index, row in m_data.iterrows():
    if row['address_address_jac_qgm_3_qgm_3']<1:
        if row['address_address_jac_qgm_3_qgm_3']<min_address_sim:
            min_address_sim=row['address_address_jac_qgm_3_qgm_3']
        print('id1:{}, id2:{}, address1:{}, address2:{}, address_address_jac_qgm_3_qgm_3:{}'.format(row['id1'], row['id2'], rest['address'][row['id1']], rest['address'][row['id2']],row['address_address_jac_qgm_3_qgm_3']))
print('min_address_sim:{}\n'.format(min_address_sim))


# Output the distinct city sim value of matched pairs
min_city_sim=1
citySim_to_cityValue={}
for index, row in m_data.iterrows():
    citySim=row['city_city_jac_qgm_3_qgm_3']
    cityValue=((rest.iloc[int(row['id1'])])['city'],(rest.iloc[int(row['id2'])])['city'])
    if citySim not in citySim_to_cityValue:
        citySim_to_cityValue[citySim]=[cityValue]
        if citySim<min_city_sim:
            min_city_sim=citySim
    else:
        if cityValue not in citySim_to_cityValue[citySim]:
            citySim_to_cityValue[citySim].append(cityValue)
print(citySim_to_cityValue)
print('min_city_sim:{}, len(citySim_to_cityValue):{}\n'.format(min_city_sim,len(citySim_to_cityValue)))

# Output the distinct flavor sim value of matched pairs
min_flavor_sim=1
flavorSim_to_flavorValue={}
for index, row in m_data.iterrows():
    flavorSim=row['flavor_flavor_jac_qgm_3_qgm_3']
    flavorValue=((rest.iloc[int(row['id1'])])['flavor'],(rest.iloc[int(row['id2'])])['flavor'])
    if flavorSim not in flavorSim_to_flavorValue:
        flavorSim_to_flavorValue[flavorSim]=[flavorValue]
        if flavorSim<min_flavor_sim:
            min_flavor_sim=flavorSim
    else:
        if flavorValue not in flavorSim_to_flavorValue[flavorSim]:
            flavorSim_to_flavorValue[flavorSim].append(flavorValue)
print(flavorSim_to_flavorValue)
print('min_flavor_sim:{}, len(flavorSim_to_flavorValue):{}\n'.format(min_flavor_sim,len(flavorSim_to_flavorValue)))
```

```python
# Explore u_data

# Output the max name sim value of unmatched pairs
max_name_sim=0
for index, row in u_data.iterrows():
    if row['name_name_jac_qgm_3_qgm_3']>max_name_sim:
        max_name_sim=row['name_name_jac_qgm_3_qgm_3']
        print('id1:{}, id2:{}, name_name_jac_qgm_3_qgm_3:{}'.format(row['id1'], row['id2'], row['name_name_jac_qgm_3_qgm_3']))
print('max_name_sim:{}\n'.format(max_name_sim))


# Output the max address sim value of unmatched pairs
max_address_sim=0
for index, row in u_data.iterrows():
    if row['address_address_jac_qgm_3_qgm_3']>max_address_sim:
        max_address_sim=row['address_address_jac_qgm_3_qgm_3']
        print('id1:{}, id2:{}, address_address_jac_qgm_3_qgm_3:{}'.format(row['id1'], row['id2'], row['address_address_jac_qgm_3_qgm_3']))
print('max_address_sim:{}\n'.format(max_address_sim))
```

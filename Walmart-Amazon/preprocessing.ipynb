```python
import sys
import py_entitymatching as em
import pandas as pd
import os
```

```python
walmart=pd.read_csv('csv_files/walmart_short.csv', dtype = {'custom_id':int})
walmart.fillna({'brand':'',
                'modelno':'',
                'title':'',
                'price':0,
                'descr':'',
                'dimensions':'',
                'shipweight':0}, inplace=True)
```

```python
amazon=pd.read_csv('csv_files/amazon_short.csv', dtype = {'custom_id':int})
amazon.fillna({'brand':'',
                'modelno':'',
                'title':'',
                'price':0,
                'descr':'',
                'dimensions':'',
                'shipweight':''}, inplace=True)
```

```python
def deal_with_brand(brand):
    return brand.lower()

def deal_with_modelno(modelno):
    tmp=[]
    useless_chars=['-','/','#']
    for char in modelno:
        if char not in useless_chars:
            tmp.append(char.lower())
    return ''.join(tmp)

def deal_with_title(title):
    return title.lower()

def deal_with_descr(descr):
    return descr.lower()

def deal_with_dimensions(dimensions):
    if dimensions=='':
        return (0,0,0)
    tmp=dimensions[:-7].split(' x ')
    tmp=[float(_) for _ in tmp]
    if len(tmp)==2:
        tmp.append(0)
    return tuple(tmp)

def deal_with_shipweight(shipweight):
    if shipweight=='':
        return 0.0
    shipweight_prefix=shipweight[:-7]
    tmp=[]
    for char in shipweight_prefix:
        if char!=' ':
            tmp.append(char)
    shipweight_prefix=''.join(tmp)
    if 'pounds' in shipweight:
        return float(shipweight_prefix)
    if 'ounces' in shipweight:
        return 0.0625*float(shipweight_prefix)
```

```python
len_walmart=len(walmart)
walmart=pd.concat([walmart,pd.DataFrame(columns=['dimensions1','dimensions2','dimensions3'])])
for i in range(len_walmart):
    walmart['brand'][i]=deal_with_brand(walmart['brand'][i])
    walmart['modelno'][i]=deal_with_modelno(walmart['modelno'][i])
    walmart['title'][i]=deal_with_title(walmart['title'][i])
    walmart['descr'][i]=deal_with_descr(walmart['descr'][i])
    dimensions=deal_with_dimensions(walmart['dimensions'][i])
    walmart['dimensions1'][i],walmart['dimensions2'][i],walmart['dimensions3'][i]=dimensions
walmart['custom_id']=walmart['custom_id'].astype('int64')
walmart.drop(columns=['dimensions'],inplace=True)
walmart.to_csv('csv_files/walmart_cleaned.csv',index=False)
```

```python
len_amazon=len(amazon)
amazon=pd.concat([amazon,pd.DataFrame(columns=['dimensions1','dimensions2','dimensions3'])])
for i in range(len_amazon):
    amazon['brand'][i]=deal_with_brand(amazon['brand'][i])
    amazon['modelno'][i]=deal_with_modelno(amazon['modelno'][i])
    amazon['title'][i]=deal_with_title(amazon['title'][i])
    amazon['descr'][i]=deal_with_descr(amazon['descr'][i])
    dimensions=deal_with_dimensions(amazon['dimensions'][i])
    if len(dimensions)>3:
        print(i,len(dimensions),amazon['title'][i])
        continue
    amazon['dimensions1'][i],amazon['dimensions2'][i],amazon['dimensions3'][i]=dimensions
    amazon['shipweight'][i]=deal_with_shipweight(amazon['shipweight'][i])
amazon['custom_id']=amazon['custom_id'].astype('int64')
amazon.drop(columns=['dimensions'],inplace=True)
amazon.to_csv('csv_files/amazon_cleaned.csv',index=False)
```

```python
# Blocking

# Overlap blocking over input tables on title: the title of matched entities should have at least 2 same words
ob=em.OverlapBlocker()
C1=ob.block_tables(A,B,'title','title',word_level=True,overlap_size=2,allow_missing=True,
                 l_output_attrs=['brand','modelno','title','price','descr','shipweight','dimensions1','dimensions2','dimensions3'],
                 r_output_attrs=['brand','modelno','title','price','descr','shipweight','dimensions1','dimensions2','dimensions3'],
                 show_progress=True)
print(len(C1))

C2=ob.block_tables(A,B,'brand','brand',word_level=True,overlap_size=1,allow_missing=True,
                 l_output_attrs=['brand','modelno','title','price','descr','shipweight','dimensions1','dimensions2','dimensions3'],
                 r_output_attrs=['brand','modelno','title','price','descr','shipweight','dimensions1','dimensions2','dimensions3'],
                 show_progress=True)
print(len(C2))

C3=em.combine_blocker_outputs_via_union([C1,C2])
print(len(C3))
```

```python
def jac_qgm_3_qgm_3(seq1, seq2):
    qgm1 = []
    qgm2 = []
    seq1 = '##' + seq1 + '$$'
    seq2 = '##' + seq2 + '$$'
    for i in range(len(seq1) - 2):
        qgm1.append(seq1[i:i + 3])
    for i in range(len(seq2) - 2):
        qgm2.append(seq2[i:i + 3])
    qgm1 = set(qgm1)
    qgm2 = set(qgm2)
    return len(qgm1 & qgm2) / len(qgm1 | qgm2)
```

```python
match=pd.read_csv('csv_files/matches_walmart_amazon.csv')
length=len(match)

columns=['brand','modelno','title','price','shipweight']
for col in columns:
    match_info=open('csv_files/match_info/{}.csv'.format(col),'w')
    for i in range(length):
        idWalmart=match['id1'][i]
        idAmazon=match['id2'][i]
        seq1=A.iloc[idWalmart-1][col]
        seq2=B.iloc[idAmazon-1][col]
        if col in ['brand','modelno','title']:
            sim=jac_qgm_3_qgm_3(seq1,seq2)
            match_info.write('{}***{}***{}\n'.format(seq1,seq2,sim))
        else:
            match_info.write('{}***{}\n'.format(seq1,seq2))
    match_info.close()
    
match_info=open('csv_files/match_info/descr.csv','w')
for i in range(length):
    idWalmart=match['id1'][i]
    idAmazon=match['id2'][i]
    seq1=A.iloc[idWalmart-1]['descr']
    seq2=B.iloc[idAmazon-1]['descr']
    match_info.write('*'+seq1+'\n')
    match_info.write('**'+seq2+'\n')
    sim=jac_qgm_3_qgm_3(seq1,seq2)
    match_info.write(str(sim)+'\n\n')
match_info.close()

columns=['dimensions1','dimensions2','dimensions2']
match_info=open('csv_files/match_info/dimensions.csv','w')
for i in range(length):
    idWalmart=match['id1'][i]
    idAmazon=match['id2'][i]
    for col in columns:
        match_info.write(str(A.iloc[idWalmart-1][col])+' ')
    match_info.write('*** ')
    for col in columns:
        match_info.write(str(B.iloc[idAmazon-1][col])+' ')
    match_info.write('\n')
match_info.close()
```

```python
import pandas as pd

# Generate product.csv by concating knowledge base/buy.csv+googleproducts.csv
def file_to_lower(file_name):
    lower_file_name=file_name[:-4]+'_lower.csv'
    file=open(file_name)
    lower_file=open(lower_file_name,'w')
    line=file.readline()
    while line:
        lower_file.write(line.lower())
        line=file.readline()
    file.close()
    lower_file.close()
    
file_to_lower('csv_files/knowledge base/best_buy.csv')
file_to_lower('csv_files/knowledge base/Buy.csv')
file_to_lower('csv_files/knowledge base/GoogleProducts.csv')

best=pd.read_csv('csv_files/knowledge base/best_buy_lower.csv')
buy=pd.read_csv('csv_files/knowledge base/Buy_lower.csv')
google=pd.read_csv('csv_files/knowledge base/GoogleProducts_lower.csv')

product=open('csv_files/product.csv','w')
product.write('title,price,descr\n')

def deal_with_price(s):
    if type(s)==float:
        return s
    tmp=[]
    for c in s:
        if c=='.' or (c>='0' and c<='9'):
            tmp.append(c)
    return float(''.join(tmp))

for index, row in best.iterrows():
    product.write('"{}",{},"{}"\n'.format(row['name'],deal_with_price(row['price']),row['description']))
for index, row in buy.iterrows():
    product.write('"{}",{},"{}"\n'.format(row['name'],deal_with_price(row['price']),row['description']))
for index, row in google.iterrows():
    product.write('"{}",{},"{}"\n'.format(row['name'],deal_with_price(row['price']),row['description']))
    
product.close()
```
